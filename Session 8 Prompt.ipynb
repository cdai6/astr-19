```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,1,0.01)

def e_1(x):
    return np.sin(x)

def e_2(x):
    return np.cos(x)
     
f,axarr = plt.subplots(1,2)
y_1 = e_1(x)
axarr[0].plot(x, y_1)
y_2 = e_2(x)
axarr[1].plot(x, y_2)
axarr[0].set_xlabel('x')
axarr[0].set_ylabel('sin(x)')
axarr[1].set_xlabel('x')
axarr[1].set_ylabel('cos(x)')
plt.show()
plt.savefig('line_plot.pdf')
```
